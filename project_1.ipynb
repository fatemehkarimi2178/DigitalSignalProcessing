```matlab
n1=0:1:20;
x1=0.5.*1i.*exp(1i.*pi/7.*n1)+3.5.*1i.*exp(-1i.*pi/7.*n1);
n2=-15:1:25;
x2=-0.5.*1i.*exp(1i.*pi/17.*n2)+0.5.*1i.*exp(-1i.*pi/17.*n2);
n3=0:1:50;
x3=(1.1.^n3).*0.7.*exp(1i.*pi/11.*n3)+(1.1.^n3).*0.7.*exp(1i.*pi/11.*3);
n4=-10:1:20;
x4=(1.1.^n4).*exp(1i.*pi/11.*n4)+(1.1.^n4).*exp(-1i.*pi/11.*n4);
c11=xcorr(x1);
c44=xcorr(x4);
c12=xcorr(x1,x2);
subplot(3,1,1);
plot(real(c11),'r');
title('autocorrolation of x1');
subplot(3,1,2);
plot(real(c44),'gr');
title('autocorrolation of x4');
subplot(3,1,3);
plot(real(c12),'b');
title('corrolation of x1,x2');
```

```matlab
clear
clc
%system input
t = 0 : 0.004 : 0.1;
f = 10; %% frequency of signal
x = 5* sin (2*pi*f*t + pi/2);
n= rand(1,length(t))-0.5;
y= n+x;
%moving average system
m1=12;
m2=12;
p=ones(m1+m2+1,1);
%system output
z=conv(y,p);
k=filter(p,1,y);
%%plot input,system,output
subplot(4,1,1);
stem(y);
title('input signal');xlabel('time index');
subplot(4,1,2);
stem(p);
title('system function');xlabel('time index')
subplot(4,1,3);
stem(z);
title('output signal');xlabel('time index')
subplot(4,1,4);
stem(k);
title('filter');xlabel('time index')
```
